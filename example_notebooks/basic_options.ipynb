#### Using simple options

Here, we demonstrate the capabilities of our options objects and show some examples that may be helpful. Our first example is to analyze the following option
* Call option on COIN
* Strike price of 52
* Maturity date of 4/14/2023

```python
import matplotlib.pyplot as plt
import numpy as np

from options.options import CallOption, PutOption
from options.enums import OptionExerciseType

from datetime import date

import warnings
warnings.filterwarnings('ignore')
```

```python
call = CallOption(
    underlying='COIN', 
    strike = 52, 
    exercise_type=OptionExerciseType.EUROPEAN, 
    expiry_date=date(2023, 4, 14)
)
```

```python
call 
```

```python
call.time_to_expiry(start_date=date.today())
```

This call option class has useful methods which help us to analyze the derivative itself

```python
S = np.arange(20, 90) # Creating a set
black_scholes_value = call.black_scholes_price(S = S, sigma = 0.6, r = 0.05)
```

```python
plt.plot(S, black_scholes_value)
plt.title("Value of Call Option using Black-Scholes") 
plt.xlabel("Stock Price")
plt.grid(axis = 'y')
plt.ylabel("Option value")
plt.show()
```
